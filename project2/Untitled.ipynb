```python
import pandas as pd
```

```python
excel = pd.read_excel('container3.xlsx')
excel.to_csv('container3.csv')
```

```python
df = pd.read_csv('container3.csv')
df.head()
```

```python
df2 = df['호출부호']
df2_unique = df2.drop_duplicates()
print(df2_unique)
```

```python
vsl = pd.read_csv('ais_for_waiting_area_vsl_info.csv')

matching_rows = vsl[vsl['CALL_SIGN'].isin(df2_unique)]

# 일치하는 데이터프레임 출력
print(matching_rows)
```

```python
matching_rows.to_csv('match_vsl3.csv', index=False)
```

```python
df3 = pd.read_csv('match_vsl3.csv')
df3.head()
```

```python
df4 = df3['CALL_SIGN'].drop_duplicates()
print(df4)
```

```python
matching_con = df[df['호출부호'].isin(df4)]

# 일치하는 데이터프레임 출력
print(matching_con)
```

```python
# 문자열을 datetime으로 변환
matching_con.loc[:, '입항일시'] = pd.to_datetime(matching_con['입항일시'])
matching_con.loc[:, '출항일시'] = pd.to_datetime(matching_con['출항일시'])

# 시간 차이 계산 (end_time - start_time)
df['time_diff'] = matching_con['출항일시'] - matching_con['입항일시']

# 결과 출력
print(df['time_diff'])
```
