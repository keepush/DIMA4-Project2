```python
import folium

# 모든 좌표를 각각의 다각형으로 지도에 그리기
polygons = [
    [{"lat": 35.07645, "lon": 129.04538}, {"lat": 35.07645, "lon": 129.03402}, {"lat": 35.06336, "lon": 129.03549}, {"lat": 35.06336, "lon": 129.06442}],
    [{"lat": 35.07645, "lon": 129.02445}, {"lat": 35.07645, "lon": 129.03112}, {"lat": 35.05281, "lon": 129.03112}, {"lat": 35.05281, "lon": 129.01497}],
    [{"lat": 35.06336, "lon": 129.06442}, {"lat": 35.06336, "lon": 129.03552}, {"lat": 35.04430, "lon": 129.03708}, {"lat": 35.04861, "lon": 129.08608}],
    [{"lat": 35.04841, "lon": 129.08572}, {"lat": 35.04430, "lon": 129.03708}, {"lat": 35.02686, "lon": 129.04391}, {"lat": 35.03731, "lon": 129.09341}],
    [{"lat": 35.02686, "lon": 129.04391}, {"lat": 35.00319, "lon": 129.04391}, {"lat": 35.00319, "lon": 129.07992}, {"lat": 35.03731, "lon": 129.09341}]
]

# 모든 좌표를 하나의 리스트로 합치기
all_coords = [point for polygon in polygons for point in polygon]

# 위도와 경도의 평균값 계산
avg_lat = sum(point["lat"] for point in all_coords) / len(all_coords)
avg_lon = sum(point["lon"] for point in all_coords) / len(all_coords)

# 계산된 중심 좌표 출력
print(f"중심 좌표는 위도: {avg_lat}, 경도: {avg_lon} 입니다.")
```

```python
# 지도 생성 (중심 좌표 설정)
m_multiple_polygons = folium.Map(location=[avg_lat, avg_lon], zoom_start=13)

# 각 다각형을 지도에 추가
for polygon in polygons:
    lat_lon_pairs = [(point["lat"], point["lon"]) for point in polygon]
    folium.Polygon(
        locations=lat_lon_pairs,
        color='blue',
        fill=True,
        fill_opacity=0.4,
    ).add_to(m_multiple_polygons)

# 지도 저장
m_multiple_polygons.save("multiple_polygons_map.html")

# 실행 후 'multiple_polygons_map.html' 파일을 열면 지도를 볼 수 있습니다.
```
