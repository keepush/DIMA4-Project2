```python
import mysql.connector

# MySQL 데이터베이스에 연결 설정
connection = mysql.connector.connect(
    host = 'localhost',
    user = 'root',
    password = 'root',
    database = 'project'
)

# 연결 확인 절차
if connection.is_connected():
    print("연결 성공")
```

```python
df = pd.read_csv('port_info.csv')

print(df.head(5))
```

```python
# MySQL 커서 생성
cursor = connection.cursor()

# 각 행을 삽입하는 SQL 쿼리 작성
for index, row in df.iterrows():
    sql = """
    INSERT INTO port (port_code, port_name, port_lat, port_lon, port_contact, port_url, port_addr, avg_working_time, avg_waiting_time)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
    ON DUPLICATE KEY UPDATE
    port_code = VALUES(port_code),
    port_name = VALUES(port_name),
    port_lat = VALUES(port_lat),
    port_lon = VALUES(port_lon),
    port_contact = VALUES(port_contact),
    port_url = VALUES(port_url),
    port_addr = VALUES(port_addr),
    avg_working_time = VALUES(avg_working_time),
    avg_waiting_time = VALUES(avg_waiting_time)
    """
    # INSERT 쿼리에 삽입할 값들
    values = (
        row['port_code'],
        row['port_name'],
        row['port_lat'],
        row['port_lon'],
        row['port_contact'],
        row['port_url'],
        row['port_addr'],
        row['avg_working_time'],
        row['avg_waiting_time']
    )
    
    cursor.execute(sql, values)

# 변경 사항을 커밋
connection.commit()

# 연결 해제
cursor.close()
connection.close()

print("CSV 데이터를 MySQL에 성공적으로 삽입했습니다.")
```
